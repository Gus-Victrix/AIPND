## Download Assets

The `download_and_unzip()` is used to download and extract the notebook assets.

```python
def download_and_unzip(url, save_path):
  print(f"Downloading and extracting assets...", end="")

  # Downloading zip file using urllib package
  urlretrieve(url, save_path)

  try:
    # Extract zip using zipfile package
    with ZipFile(save_path) as z:
      z.extractall(os.path.split(save_path)[0])
    print("Done")
  except Exception as e:
    print("\nInvalid file:", e)
```

```python
URL = r"https://www.dropbox.com/s/qhhlqcica1nvtaw/opencv_bootcamp_assets_NB1.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB1.zip")

# Download if Zip does not exist
if not os.path.exists(asset_zip_path):
  download_and_unzip(URL, asset_zip_path)
```

## Splitting and Merging Coor Channels

- `split()`: Divides multi-channel array into several single-channel arrays.
- `merge()`: Merges several arrays to make single multi-channel array.

```python
# Split Image into BGR components
img_NZ = cv.imread("New_Zealand_Lake.jpg", cv.IMREAD_COLOR)

b, g, r = cv.split(img_NZ)

# Show channels
plt.figure(figsize=[20, 51])

plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel");
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel");
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel");

#  Merge the individual channels into a BGR image
img_merged = cv.merge((b, g, r))
# Show merged output
plt.subplot(144);plt.imshow(img_merged[:, :, ::-1]);plt.title("Merged Output")
```

## Converting to different channels

- Use `cvtColor()`

### Changing from BGR to RGB

```python
img_NZ_rgb = cv.cvtColor(img_NZ, cv.COLOR_BGR2RGB)
stuff = plt.imshow(img_NZ_rgb)
```

## Changing to HSV color space

```python
img_hsv = cv.cvtColor(img_NZ, cv.COLOR_BGR2HSV)

# Split into channels
h, s, v = cv.split(img_hsv)

# Show the channel values
plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(h, cmap="gray");plt.title("Hue Channel")
plt.subplot(142);plt.imshow(s, cmap="gray");plt.title("Saturation Channel")
plt.subplot(143);plt.imshow(v, cmap="gray");plt.title("Value Channel")
plt.subplot(144);plt.imshow(img_NZ_rgb);plt.title("Original Image");
plt.show()

# Modify the Value
new_v = v + 5
# Merge the new and display
new_img_hsv = cv.merge([h, s, new_v])
# Turn to RGB
new_img = cv.cvtColor(new_img_hsv, cv.COLOR_HSV2RGB)
# Display new image
plt.imshow(new_img);plt.title("Increased Value")
```
